```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
function_name = 'sine_10pi'
start = 0
stop = 10 * np.pi
number_of_points = [3, 9, 27, 81]

def my_function(x):
    return np.sin(x)
```

```python
for N in number_of_points:
    # Sample values
    x = np.linspace(start, stop, N) # equally-spaced points
    y = my_function(x) # the only point where we know the actual function
    
    # Save data
    data = pd.DataFrame(data = {"x": x, "y": y})
    data.to_csv(f"../training_data/input_{function_name}_linspace_{N}.csv", index = False) # save to csv

    # Represent figure
    plt.figure(figsize=(2, 2))
    plt.plot(x, y)
    plt.scatter(x, y, s = 16, c = "red")
    plt.title(f"{function_name} - {N} points")
    plt.show()
```

```python
for N in [30, 90, 270, 1000, 10000]:
    # Sample values
    x_training = np.linspace(start, stop, N) # equally-spaced points in [0, 2]
    x = x_training + (x_training[1] - x_training[0]) / 2 # shift training data
    x = x[:-1] # trim outlier
    y = my_function(x) # the only point where we know the actual function
    
    # Save data
    data = pd.DataFrame(data = {"x": x, "y": y})
    data.to_csv(f"../verification_data/verification_{function_name}_linspace_{N}.csv", index = False) # save to csv

    # Represent figure
    plt.figure(figsize=(2, 2))
    plt.plot(x, y)
    plt.scatter(x, y, s = 16, c = "red")
    plt.title(f"{function_name} - {N-1} points")
    plt.show()
```
