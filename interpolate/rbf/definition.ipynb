```python
import sympy
```

```python
def rbf_function(eps: sympy.Symbol, r: sympy.Symbol):
    return sympy.sqrt(1 + eps**2 * r**2)
```

```python
r = sympy.Symbol('r')
p1 = sympy.plot(rbf_function(0.1, r), show=False, label="eps=0.1")
p1.append(sympy.plot(rbf_function(0.5, r), show=False, label="eps=0.5")[0])
p1.append(sympy.plot(rbf_function(1, r), show=False, label="eps=1")[0])
p1.legend = True
p1.show()
```

```python
from sympy.plotting import plot3d

eps = sympy.Symbol('eps')
plot3d(rbf_function(eps, r))
```
