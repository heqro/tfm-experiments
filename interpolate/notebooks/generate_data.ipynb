```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
function_name = 'exp_inv'
```

```python
for N in [3, 5, 10, 15]:
    # Sample values
    x = np.linspace(start = 0, stop = 5, num = N) # equally-spaced points in [0, 2]
    y = 1 / np.exp(x) # the only point where we know the actual function
    
    # Save data
    data = pd.DataFrame(data = {"x": x, "y": y})
    data.to_csv(f"../training_data/input_{function_name}_linspace_{N}.csv", index = False) # save to csv

    # Represent figure
    plt.figure(figsize=(2, 2))
    plt.plot(x, y)
    plt.scatter(x, y, s = 16, c = "red")
    plt.title(f"{function_name} - {N} points")
    plt.show()
```

```python
for N in [30, 90, 270, 1000, 10000]:
    # Sample values
    x_training = np.linspace(start = 0, stop = np.pi, num = N) # equally-spaced points in [0, 2]
    x = x_training + (x_training[1] - x_training[0]) / 2 # shift training data
    x = x[:-1] # trim outlier
    y = 1 / np.exp(x) # the only point where we know the actual function
    
    # Save data
    data = pd.DataFrame(data = {"x": x, "y": y})
    data.to_csv(f"../verification_data/verification_{function_name}_linspace_{N}.csv", index = False) # save to csv

    # Represent figure
    plt.figure(figsize=(2, 2))
    plt.plot(x, y)
    plt.scatter(x, y, s = 16, c = "red")
    plt.title(f"{function_name} - {N-1} points")
    plt.show()
```
